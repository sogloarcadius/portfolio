# description_en

    ## Context
 
     During my training leading to the computer networks and telecommunications administrator, the students of the second year have to do a tutored project proposed by the teachers of the IUT (Technology Institute). I have worked in group with another student on the implementation of a mail server on Linux.

    ## Requirements
 
     *  Install and configure a virtual environment running Linux : Ubuntu (14.04 LTS)
     *  Install and configure a SMTP server (Postfix)
     *  Install and configure an anti-spam (Spamassasin) and antivirus (Clam AV)
     *  Install and configure a POP3 / IMAP4 server (Dovecot)
     *  Install and configure a Webmail (Roundcube)
     *  Install and configure a mail client (Thunderbird)
     *  Analyse the secure e-mail exchange mechanism
     *  Use Wireshark to check that the services are performing as expected
     *  Create Installation guide for each service


    ## Project Report (Installation Guide)

# description_es

    ## Objetivo

    Para obtener mi diploma de redes de las computadoras y telecomunicaciones, tenia que hacer un proyecto supervisado por los profesores del instituto universitario de technologia (IUT). En mi caso, trabaje en conjunto con otro alumno sobre la implementacion de un servidor de coreo electronico con el sistema operativo Linux.

    ## Requisitos
    
    * Instalar y configurar un servidor virtual con Linux : Ubuntu (14.04 LTS)
    * Instalar y configurar un servidor SMTP (Postfix)
    * Instalar y configurar un anti-spam (Spamassasin) y antivirus (Clam AV)
    * Instalar y configurar un servidor POP3 / IMAP4 (Dovecot)
    * Instalar y configurar un Webmail (Roundcube)
    * Instalar y configurar un cliente de correo (Thunderbird)
    * Analizar el mecanismo de securidad de intercambio de correo electrónico
    * Utilice Wireshark para comprobar que los servicios están funcionando
    * Crear una guía de instalación para cada servicio
