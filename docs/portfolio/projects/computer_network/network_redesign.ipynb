# description_en
    
    ## Context

    During my master degree courses at ENSSAT, I setup a computer network with another classmates. The goal of the project is to reuse our skills in computer network on a real use case. We are suppose to redesign a company network.
    
    The project consist of practising some skills of a network engineer by analysing a real use case requirements and set up a solution to satisfy the requirements (Computer network architecture)
    
    
    The following fonctionalities and services have been implemented in the final architecture : 
    
    * Vlans, VOIP, DNS, DHCP
    * SFTP et SSH
    * Firewall
    * Mail server
    * QoS
    
    The following equipments have been used :
    
    * cisco router 2811
    * Cisco switch  2960
    * IP Phone Alcatel

# description_fr
    
    ## Contexte
    
    Dans le cadre de ma formation d'ingénieur à l'ENSSAT, j'ai été amené à réaliser un projet de réseaux en équipe.
    L'objectif du projet est de réutiliser les connaissances acquises en réseaux sur un cas d'usage entreprise.
    Il s'agit de la modification de la haute disponibilité du réseau d'entreprise suite à la fusion de deux sociétés( NITFLEX & ANI).
     
    Le projet à permis de mettre en oeuvre les compétences d'un ingénieur réseau qui consiste à analyser un cahier des charges puis proposer une solution pour répondre aux attentes (conception de l'architecture réseau).
    
    Les fonctionnalités et services suivants sont à présent disponible dans la nouvelle configuration du réseau :
    
    * Vlans, VOIP, DNS, DHCP
    * SFTP et SSH
    * Pare-feu - filtrage
    * Serveur mail
    * QoS
    
    Les équipements suivants ont été utilisés :
    
    * Routeur cisco 2811
    * Commutateur Cisco  2960
    * Téléphone IP Alcatel
