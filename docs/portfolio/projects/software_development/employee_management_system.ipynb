# description_en

    ## Context
    
    This project was done as part of the JAVA module of my master. The goal is to create a UML diagram and the java code based on the requirements. In my case the requirement is to setup a java application to manage the employees, the projects and the tasks.
    
    The java application should allow users to add, view and delete new employees, customers, projects and tasks.
    
    The application should allow the tracking of the employees tasks.
    
    The tools needed to develop the application are :
    
    * JDK 1.8.0 
    * Eclipse IDE

# description_fr

    ## Contexte
    
    Pour mettre en pratique mes connaissances en JAVA, J'ai été amené à réaliser un projet en monôme. 
    L’objectif est de mettre en œuvre un modèle UML pour répondre à un cahier des charges et ensuite coder celui-ci dans le langage JAVA.

    Le cahier des charges est de réaliser une application pour la gestion des taches des employées pour suivre les projets d’une entreprise. Cette application offre un nombre de fonctionnalités de base à savoir la consultation, l’ajout, la modification et la suppression d’un projet, d’un client, d’un employé ou d’une tache.

    Les outils utilisés pour mettre en œuvre le système de gestion de taches sont :

    * JDK 1.8.0
    * IDE Eclipse
