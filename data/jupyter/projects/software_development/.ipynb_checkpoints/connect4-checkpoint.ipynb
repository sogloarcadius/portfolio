# description_en

    ## Context
    
    This project was done as part of my master degree courses at ENSSAT. To practise the java skills, we should develop a clone of the connect four game. The idea is to create the game using SWING and JAVA2D. The idea is to create the game using SWING and the MVC design pattern. The game should respect some constraints :
    
    * Game board customizable
    * User interaction : username/pseudo
    * Display of best scores of previous games (saved in file)
    
    The tools used to develop the game are :
    
    * JDK 1.8.0
    * SWING API
    * Java2D API
    * Eclipse IDE

# description_fr

    ## Contexte

    Pour mettre en pratique mes connaissances acquises en programmation java, j'ai été amené à réaliser un projet en monôme. L’objectif est de réaliser un clone du jeu puissance 4 en utilisant le modèle MVC (Modèle Vue Contrôleur) et la librairie Swing.
    
    Le cahier des charges est de réaliser le clone du jeu en respectant certaines contraintes :

    * Taille du plateau configurable
    * Interaction avec l’utilisateur : saisie des noms/pseudos des joueurs
    * Affichage des meilleurs scores (et sauvegarde dans un fichier).

    Les outils utilisés pour mettre en œuvre le clone du jeu puissance 4 sont :
    
    * JDK 1.8.0
    * Librairie Swing
    * Librairie Java2D 
    * IDE Eclipse
