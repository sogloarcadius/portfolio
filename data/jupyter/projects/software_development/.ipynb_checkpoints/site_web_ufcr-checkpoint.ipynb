# description_en

    ## Context

      When I was looking for an Internship for 2 Years Degree in Computer Systems Networking and Telecommunication Diploma DUT, I applied for a developer position to set up an association's website.
    Unfortunately, the association does not have the legal status to assert the mission as an internship subject, however I have agreed to set up the website during my free time. I set up the first version of the website in 2015.

    ## Tasks
 
    * Website specifications with the client
    * Choice of hosting company
    * Purchase of domain name
    * Creation of the website design
    * Development of the website with Wordpress
    * Member space creation
 
    [Link to the website](http://parti-ufcr.fr/mentions-legales)


# description_fr

    ## Contexte

    Lors de ma recherche de stage de fin de formation de DUT Réseaux et Télécommunications, j'ai postulé sur une annonce de recherche de développeur pour mettre en place le site web d'une association.
    Malheuresement, l'association demandeuse n'a pas le statut juridique nécessaire pour faire valoir la mission comme un sujet de stage, cependant j'ai accepté de mettre en place le site web pendant mon temps libre.
    J'ai mis en place la premiere version du site web en 2015. 

    ## Cahier des charges en collaboration avec le responsable de l'association : 

    *  Élaboration du cahier des charges du site avec le client
    *  Choix du type d'hébergement internet
    *  Achat du nom de domaine et du type d'hébergement
    *  Création de la maquette du site 
    *  Création du site avec Wordpress et ajout des différentes fonctionnalités comme élaborés dans le cahier des charges
    *  Mise en place de l'espace membres
    
    [Lien vers le site internet](http://parti-ufcr.fr/mentions-legales)


# description_es

    ## Contexto

    Cuando buscaba una empresa para hacer mis praticas para obtener mi titulo  de administrador de las redes de las computadoras y Telecomunicaciones (DUT), mandé mi aplicacion a una asociacion para crear su sitio web.
    La organización solicitante no tenia la capacidad jurídica necesaria para validar el trabajo con la universidad, sin embargo, me puse de acuerdo con la asociacion para crear su sitio web durante mi tiempo libre.
    He publicado la primera versión del sitio web en 2015.
   
    ## Requisitos 
    
    *  Especificaciones del sitio web con el cliente
    *  Selecion de la empresa de alojamiento
    *  Compra de nombre de dominio
    *  Creación de la maqueta del sitio web 
    *  Desarollo del sitio web con Wordpress
    *  Creación del espacio para miembros
    
    [enlace del Sitio web](http://parti-ufcr.fr/mentions-legales)
