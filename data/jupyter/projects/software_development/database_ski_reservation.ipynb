# description_en

    ## Context
    
    This project was done as part of the database module of my master. It was a group project and I did the project with another classmate. The goal is to setup a database for the ski reservation system.
    
    The database have to allow the reservation of at least fourty appartments of differents type in three different places. We are expected to setup the following use cases :
    
    * Registration or login of the customer
    * Booking of an appartment
    * Confirmation of booking with the payement of the deposit
    * Automated deletion of booking not confirmed within 15 days.
    
    During the project we follow the classic workflow of database creation : 
    
    * Conceptual and logical modeling
    * Physical modeling
    * Data Definition Language (SQL code, indexes, triggers, procedures, functions)
    * Testing with data
    
    The tools used are :
    
    * RDBMS : MySQL
    * SQL client : MySQL Workbench

# description_fr

    ## Contexte 
    
    Dans le cadre du module de base de données, j'ai été amené à réaliser un projet en binôme. L’objectif est de réaliser une base de données pour la gestion de location des appartements d’une station de sports d’hiver.
       
    La base de données à créer doit permettre de gérer un système de location d'une quarantaine d'appartements de catégories différentes et reparties dans trois sites.
    
    Nous devrons mettre en place des mécanismes pour permettre :

    * L'inscription ou la connexion d'un client,
    * La demande de réservation d'un appartement,
    * La validation de la réservation avec le paiement des arrhes,
    * La suppression automatique des réservations de plus de 15 jours
    
    Nous avons réaliser toutes les phases du cycle de création d'une base de données : 

    * Model logique
    * Model physique
    * Ecriture du code SQL (indexes, triggers, fonctions, procédures)
    * Tests des contraintes avec des jeu de données 

    Les outils utilisés sont :
    
    * SRGBD : MySQL
    * Client SQL : MySQL Workbench

# description_es

    ## Contexto
    
    Este proyecto fue en grupo de dos durante mis clases de base de datos del master. El objetivo es de crear una base de datos para la reservacion de apartamentos en una estacion de ski.
    
    La base de datos debe permitir la reservacion de quarenta apartamentos en tres diferentes sitios. 
    
    El systema debe manejar la:
    
    * Inscripcion o connexion de los clientes
    * Reservacion de un apartamento
    * Confirmacion de una reservacion cuando el cliente paga el deposito  
    * Suprecion automatica de las revervaciones no confirmada en 15 dias.
    
    Las etapas siguentes fueron respectadas para crear la base de datos :
    
    * Modelo logico 
    * Modelo fisico 
    * Codigo SQL (indexes, triggers, proceduras, funciones)
    * Test del sistema
    
    Las herramientas utilizadas son :
    
    * SRGBD : MySQL
    * Cliente SQL : MySQL WorkBench
