# description_en
    
    ## Context
    
    This project was done as part of the web application development module of my master courses. The idea is to create a web application to manage a Judo Tournament using the PHP framework CodeIgniter and the MVC design pattern.
    
    We develop the following use cases :
    
    * Users management
    * Competition management
    * Fights management
    
    We use the PHP framework CodeIgniter and the relational database management system PostgreSQL to store the data. A focus on the project management and team management was required and evaluated during this project. Another constraints was the time management.

# description_fr

    ## Contexte
    
    Dans le cadre de ma formation d"ingénieur j'ai été amené à réaliser une application Web en équipe pour la gestion des tournois de judo. Les contraintes majeures pour ce projet sont l’utilisation du **Framework CodeIgniter** et le respect du modèle Modèle Vue Controller. 
    
    Nous avons développés les fonctionnalités suivantes :
    
    * Gestion des utilisateurs
    * Gestion du tournoi
    * Gestion des combats
    
    Nous avons utilisé le framework CodeIgniter et le système de gestion de base de données PostgreSQL pour le stockage des données. Un point également important dans ce projet est l’organisation de l'équipe pour réaliser le travail demandé dans les délais imposés.

    * Gestiòn de los combates
