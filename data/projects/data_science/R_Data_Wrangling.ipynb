# description_en

    ## Context
    
    This project was done as part of the data analytics module during my Erasmus Exchange at DIT, Ireland. During  this project, I reuse my **R** skills on tidying datasets, transforming and visualizing data.
    
    The project consist of applying the data wrangling techniques in real world datasets. The tasks have covered data generated by the activities of the citizens of Dublin in the course of their daily pursuits :
    
    * Quarterly Economic Indicators
    * Real-time bike Info
    * Dublin Bus 
    * Footfall pattern
    
    With R, I import the differents datasets and apply data analysis techniques on them.
